```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = lambda x: 100 - 2 * x
f2 = lambda x: 80 - x
x = np.linspace(0, 110, num = 10000)
```

```python
fig, axs = plt.subplots(figsize = (10,6), dpi = 100)
axs.plot(x, f1(x), color = 'red', label = '2x + y = 100')
axs.plot(x, f2(x), color = 'blue',label = 'x + y = 80')
axs.scatter([0, 20, x[4544]], [80, 60, 0], color = 'green')
axs.set_xlim(0,100)
axs.set_ylim(0,125)
axs.legend()
axs.grid(True)
fig.tight_layout()
plt.show()
```
