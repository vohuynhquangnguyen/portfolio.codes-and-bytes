a. 
$H_0: \mu = 25, H_1: \mu ≠ 25$. 

This is the two-tailed test, the test statistic and its corresponding p-value are computed as:
* If the population variance is known (the test statistic follows the standard normal distribution): 
$$
\begin{aligned}
Z_0 &= \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}} \\
p &= 2\Big(1 - \Phi(|Z_0|) \Big)
\end{aligned}
$$
* If the population variance is unknown (the test statistic follows the t-distribution with $n-1$ degree of freedom): 
$$
\begin{aligned}
T_0 &= \frac{\bar{x} - \mu_0}{s/\sqrt{n}} \\
p &= 2\Big(1 - \Phi(|T_0|) \Big)
\end{aligned}
$$

b. $H_0: \sigma > 10, H_1: \sigma = 10$

The correct hypotheses should be: $H_0: \sigma^2 = 10, H_1: \sigma^2 > 10$. Consider the corrected hypotheses:
* This is the upper-tailed test.
* The test statistic, which follows the Chi-square distribution with $n-1$ degree of freedom, and the corresponding p-value are computed as:
$$
\begin{aligned}
\chi^2_0 &= \frac{(n-1)s^2}{\sigma_0^2} \\
p &= 1 - \Phi(\chi^2_0)
\end{aligned}
$$

c. 
$H_0∶ \bar{x} = 50, H_1∶ \bar{x} ≠ 50$ 

The correct hypotheses should be: $H_0: \mu = 50, H_1: \mu \neq 50$. Consider the corrected hypotheses:

* The test statistic, which follows the standard normal distribution, and the corresponding p-value are computed as:
$$
\begin{aligned}

* The sample size is $n=4$.
* The hypothesis test: $H_0:\mu = 12$ versus $H_a: \mu<12$. This is lower-tailed test.

Recall that Type I error is the probability of rejecting $H_0$ when $H_0$ is True. Hence, the Type I error probability if the critical region is defined as $\bar{x} < 11.5$ is given as:
$$
\alpha = P(\bar{X}<11.5|\mu = 12) \equiv P(Z<\frac{11.5-12}{0.5/\sqrt{4}}) = P(Z\leq-2)\approx 0.0227
$$

b.

Recall that Type II error is the probability of failing to reject $H_0$ when $H_0$ is False. Hence, the Type II error probability if (1) the critical region is $\bar{x} < 11.5$ and (2) the true mean is 11.25 is given as:
$$
\beta = P(\bar{X}>11.5|\mu = 11.25) \equiv P(Z\geq\frac{11.5-11.25}{0.5/\sqrt{4}}) = P(Z\geq 1)\approx 0.1586
$$

c.

The Type II error probability if (1) the critical region is $\bar{x} < 11.5$ and (2) the true mean is 11.5 is given as:
$$
\beta = P(\bar{X}>11.5|\mu = 11.25) \equiv P(Z\geq\frac{11.5-11.5}{0.5/\sqrt{4}}) = P(Z\geq 0)= 0.5
