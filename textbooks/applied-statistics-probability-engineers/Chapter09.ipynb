#### Problem 9.1.1
a. 
$H_0: \mu = 25, H_1: \mu ≠ 25$. 

This is the two-tailed test, the test statistic and its corresponding p-value are computed as:
* If the population variance is known (the test statistic follows the standard normal distribution): 
$$
\begin{aligned}
Z_0 &= \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}} \\
p &= 2\Big(1 - \Phi(|Z_0|) \Big)
\end{aligned}
$$
* If the population variance is unknown (the test statistic follows the t-distribution with $n-1$ degree of freedom): 
$$
\begin{aligned}
T_0 &= \frac{\bar{x} - \mu_0}{s/\sqrt{n}} \\
p &= 2\Big(1 - \Phi(|T_0|) \Big)
\end{aligned}
$$

b. $H_0: \sigma > 10, H_1: \sigma = 10$

The correct hypotheses should be: $H_0: \sigma^2 = 10, H_1: \sigma^2 > 10$. Consider the corrected hypotheses:
* This is the upper-tailed test.
* The test statistic, which follows the Chi-square distribution with $n-1$ degree of freedom, and the corresponding p-value are computed as:
$$
\begin{aligned}
\chi^2_0 &= \frac{(n-1)s^2}{\sigma_0^2} \\
p &= 1 - \Phi(\chi^2_0)
\end{aligned}
$$

c. 
$H_0∶ \bar{x} = 50, H_1∶ \bar{x} ≠ 50$ 

The correct hypotheses should be: $H_0: \mu = 50, H_1: \mu \neq 50$. Consider the corrected hypotheses:
* This is the two-tailed test.
* The test statistic, which follows the standard normal distribution, and the corresponding p-value are computed as:
$$
\begin{aligned}
Z_0 &= \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}} \\
p &= 2\Big(1 - \Phi(|Z_0|) \Big)
\end{aligned}
$$


d. 
$H_0: p = 0.1, H1: p = 0.5$

The correct hypotheses should be: $H_0: p = 0.1, H_1: p \neq 0.1$. Consider the corrected hypotheses:
* This is the two-tailed test.
* The test statistic, which follows the standard normal distribution, and the corresponding p-value are computed as:
$$
\begin{aligned}
Z_0 &= \frac{\hat{p} - p_0}{\sqrt{p_0(1-p_0)/n}} \\
p &= 2(1-\Phi(|Z_0|))
\end{aligned}
$$

e. 
$H_0: s = 30, H_1: s > 30$

The correct hypotheses should be: $H_0: \sigma^2 = 30^2, H_1: \sigma^2 > 30^2$. Consider the corrected hypotheses:
* This is the upper-tailed test.
* The test statistic, which follows the Chi-square distribution with $n-1$ degree of freedom, and the corresponding p-value are computed as:
$$
\begin{aligned}
\chi^2_0 &= \frac{(n-1)s^2}{\sigma_0^2} \\
p &= 1 - \Phi(\chi^2_0)
\end{aligned}
$$

### Problem 9.1.3
Given that:
* The claimed mean thread and the standard deviation are $\mu_0 = 12$ and $\sigma_0=0.5$, respectively.
* The sample size is $n=4$.
* The hypothesis test: $H_0:\mu = 12$ versus $H_a: \mu<12$. This is lower-tailed test.

a.

Recall that Type I error is the probability of rejecting $H_0$ when $H_0$ is True. Hence, the Type I error probability if the critical region is defined as $\bar{x} < 11.5$ is given as:
$$
\alpha = P(\bar{X}<11.5|\mu = 12) \equiv P(Z<\frac{11.5-12}{0.5/\sqrt{4}}) = P(Z\leq-2)\approx 0.0227
$$

b.

Recall that Type II error is the probability of failing to reject $H_0$ when $H_0$ is False. Hence, the Type II error probability if (1) the critical region is $\bar{x} < 11.5$ and (2) the true mean is 11.25 is given as:
$$
\beta = P(\bar{X}>11.5|\mu = 11.25) \equiv P(Z\geq\frac{11.5-11.25}{0.5/\sqrt{4}}) = P(Z\geq 1)\approx 0.1586
$$

c.

The Type II error probability if (1) the critical region is $\bar{x} < 11.5$ and (2) the true mean is 11.5 is given as:
$$
\beta = P(\bar{X}>11.5|\mu = 11.25) \equiv P(Z\geq\frac{11.5-11.5}{0.5/\sqrt{4}}) = P(Z\geq 0)= 0.5
$$

### Problem 9.1.10
Given that:
* The output voltage is assumed to be normally distributed with standard deviation 0.25 volt.
* $H_0: \mu = 5$ versus $H_1: \mu \neq 5$. 
* The sample size is $n = 8$.

a. 

Given that the acceptance region is $4.85 \leq \bar{x} \leq 5.15$, the value of $\alpha$ is given as:
$$
\alpha = 1 - P(4.85 \leq x \leq 5.15) = 1 - P(\frac{4.85 - 5}{0.25/\sqrt{8}}\leq\bar{x}\leq\frac{5.15 - 5}{0.25/\sqrt{8}}) \approx 1 - 0.9103 = 0.08970
$$

b. 

The power of the test for detecting a true mean output voltage of 5.1 volts is given as:
$$
\begin{aligned}
\text{power} &= 1 - \beta = 1 - \Phi(Z_{\alpha/2}-Z_{\beta,u}) - \Phi(-Z_{\alpha/2}-Z_{\beta,l}) \\
&= 1 - \Phi(1.696-\frac{5.15-5.1}{0.25/\sqrt{8}}) - \Phi(-1.696-\frac{4.85-5.1}{0.25/\sqrt{8}}) \\
&= 1 - \Phi(1.13031) - \Phi(1.13243) = 1 - 0.1291 - 0.1287 = 0.74220
\end{aligned}
$$

### Problem 9.1.11
Given that:
* $H_0: p = 0.4$ versus $H_a: p \neq 0.4$.
* The sample size is $n=15$.
* The acceptance region is $A: \frac{4}{15}\leq\hat{p}\leq\frac{8}{15}$

a. 

The type I error probability for this procedure, assuming that $p = 0.4$ is given as:
$$
\begin{aligned}
P(\hat{p}\notin A|p = 0.4) &= 1 - P(\frac{4}{15}\leq\hat{p}\leq\frac{8}{15}) \\
&= 1 - P\Big(\frac{4/15 - 0.4}{\sqrt{0.4(1-0.4)/15}}\Big)\leq\hat{p}\leq P\Big(\frac{8/15 - 0.4}{\sqrt{0.4(1-0.4)/15}}\Big) \\
&\approx 1 - P(-1.05409\leq\hat{p}\leq 1.05409) \\
&\approx 1 - 0.7801 = 0.21990
\end{aligned}
$$

b. 

The probability of committing a type II error if the true proportion is really $p = 0.2$.
$$
\begin{aligned}
\beta = P(\hat{p}\in A|p = 0.2) &= \Phi(Z_{\alpha/2} - Z_{\beta,u}) - \Phi(-Z_{\alpha/2} - Z_{\beta,l}) \\
&= \Phi(1.226 - \frac{8/15 - 0.2}{\sqrt{0.4(1-0.4)/15}}) - \Phi(-1.226 - \frac{4/15 - 0.2}{\sqrt{0.4(1-0.4)/15}}) \\
&\approx \Phi(-1.40923) - \Phi(-1.75303) \approx 0.0395\\

### Problem 9.2.6 
Given that:
* The sample size is  $n = 15$, the sample mean is $\bar{x} = 2.78$.
* The population is normally distributed with $\sigma = 0.9$.

a. 

The hypothesis test is $H_0: \mu = 3$ versus $H_1: \mu \neq 3$ with $\alpha = 0.05$, which is the two-tailed test. We can conduct the test as follows:
* Compute the test statistic: $Z_0 = \frac{2.78-3}{0.9/\sqrt{15}}= -0.4673$
* Evaluate rejection criteria: the critical value is $Z_{\alpha/2} =1.959$. since $|Z_0| < Z_{\alpha/2}$, the result of the test is that we fail to reject $H_0$.

b. 

The power of this test if $\mu = 3.25$ is given as:
$$
\text{power} = 1 - \beta = 1 - \Phi(1.959 - \frac{3.25-3}{0.9/\sqrt{15}}) - \Phi(-1.959 - \frac{3.25-3}{0.9/\sqrt{15}}) \approx 1 - 0.8102 = 0.18980
$$

c. 

The sample size would be required to detect a true mean of 3.75 if we wanted the power to be at least 0.9 is given as:
$$
\beta = 1 - 0.9 = 0.1 \rightarrow n \simeq \frac{(1.959 + 1.281)^2 \times0.9^2}{(3.75 - 3)^2} \approx 16
$$

b. 

The P-value for the test in part (a) is given as:
$$
p = 1 - \Phi(Z_0) = 0.1029
$$
