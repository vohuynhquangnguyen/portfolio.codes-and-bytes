## Exercise 51: Roots of a Quadratic Function

are constants, and $a$ is non-zero. Its roots can be identiﬁed by ﬁnding the values of $x$
that satisfy the quadratic equation $ax^2 + bx + c = 0$. These values can be computed
using the quadratic formula, shown below. A quadratic function may have 0, 1 or 2
real roots.
$$
\text{root} = \frac{−b \pm \sqrt{b^2 −4ac}}{2a}
$$
The portion of the expression under the square root sign is called the discriminant.
If the discriminant is negative then the quadratic equation does not have any real
roots. If the discriminant is 0, then the equation has one real root. Otherwise the
equation has two real roots, and the expression must be evaluated twice, once using
a plus sign, and once using a minus sign, when computing the numerator.

Write a program that computes the real roots of a quadratic function. Your program should begin by prompting the user for the values of $a$, $b$ and $c$. Then it should
display a message indicating the number of real roots, along with the values of the
real roots (if any).

```python
##
# EXERCISE 51:
#
import math

a, b, c = [int (x) for x in input('Please enter the coefficients a, b, and c to solve the quadratic function: ').split()]
a = int(a)
b = int(b)
c = int(c)

if (a == 0):
    print('The coefficient a must have a non-zero value.')
else:
    discriminant = b ** 2 - 4 * a * c
    if (discriminant < 0):
        print('This quadratic equation does not have any real root.')
    elif (discriminant > 0):
        print('This quadractic equation has two real roots.')
        root_1 = (-1 * b + math.sqrt(discriminant)) / (2 * a)
        root_2 = (-1 * b - math.sqrt(discriminant)) / (2 * a)
        print(f'First root: {root_1} - Second root: {root_2}')
    elif (discriminant == 0):
        print('This quadratic equation has only one real root.')
        root = -1 * b / (2 * a)
        print(f'Root: {root}')
```

for $15.00 a month. Each additional minute of air time costs $0.25, while additional
text messages cost $0.15 each. All cell phone bills include an additional charge of
$0.44 to support 911 call centers, and the entire bill (including the 911 charge) is
subject to 5 percent sales tax.
Write a program that reads the number of minutes and text messages used in a
month from the user. Display the base charge, additional minutes charge (if any),
additional text message charge (if any), the 911 fee, tax and total bill amount. Only
display the additional minute and text message charges if the user incurred costs in
these categories. Ensure that all of the charges are displayed using 2 decimal places.

```python
##
# EXERCISE 57:
#
original_cost = 15.00
cost_additional_airtime = 0.25
cost_additional_texts = 0.15
cost_911 = 0.44
tax_rate = 0.05

n_of_minutes = int(input('Please enter the number of minutes used in a month: '))
n_of_texts = int(input('Please enter the number of text messages used in a month: '))

print(f'Base Charge: {original_cost: .2f}')
if (n_of_minutes <= 50) and (n_of_texts <= 50):
    total_bill_amount = (original_cost + cost_911) * (1 + tax_rate)
else:
    additional_airtime = n_of_minutes - 50
    additional_airtime_charge = additional_airtime * 0.25
    if (additional_airtime_charge > 0):
        print(f'Additional Minutes Charge: {additional_airtime_charge: .2f}')

    additional_texts = n_of_texts - 50
    additional_texts_charge = additional_texts * 0.15
    if (additional_texts_charge > 0):
        print(f'Additional Text Message Charge: {additional_texts_charge: .2f}')
    
    total_bill_amount = (original_cost + cost_911 + additional_airtime_charge + additional_texts_charge) * (1 + tax_rate)

print(f'911 Fee: {cost_911: .2f}')
print(f'Tax: {tax_rate: .2f}')
print(f'Total Bill Amount: {total_bill_amount: .2f}')
```
